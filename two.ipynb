##Median Salary vs Data Science Jobs

```python
# Install datasets Library (if not already installed)
# !pip install datasets

# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  # we will use this for plotting

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.groupby('job_title_short')['salary_year_avg'].median()
```

```python
df.groupby('job_title_short')['salary_year_avg'].median().sort_values()
```

```python
df.groupby('job_title_short')['salary_year_avg'].mean().sort_values()
```

```python
job_salary = df.groupby('job_title_short')['salary_year_avg'].mean().sort_values()
```

```python
job_salary.plot(kind='barh')
plt.xlabel('Salary ($USD)')
plt.ylabel('')
plt.title('Median Salary by Job')
```
